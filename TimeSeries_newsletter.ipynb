#Analisando Dados de uma NewsLetter

```python
import pandas as pd;
import seaborn as sns;
import matplotlib.pyplot as plt;
import numpy as py;

assinantes = pd.read_csv('newsletter_alucar.csv')
assinantes.head(10)
```

```python
#Confirmando os dados e alterando o tipo do mês de objeto para date

assinantes['mes'] = pd.to_datetime(assinantes['mes'])

print(f'Quantidade total de linhas e colunas: {assinantes.shape}')
print(f'Quantidade total de dados Nulos: {assinantes.isna().sum().sum()}')
```

```python
#Conferindo o "Diff" entre o aumento de assinantes e a aceleração deles

assinantes['aumento'] = assinantes['assinantes'].diff()

assinantes['aceleracao'] = assinantes['aumento'].diff()
```

```python
#Criando a função de comparação 

def plot_comparacao(x,y1,y2,y3, database, titulo):
  plt.figure(figsize=(12,6))
  grafico = plt.subplot(3,1,1)
  grafico.set_title(titulo,fontsize=18, loc='left')
  sns.lineplot(x=x, y=y1, data=database)

  plt.subplot(3,1,2)
  sns.lineplot(x=x, y=y2, data=database)

  plt.subplot(3,1,3)
  sns.lineplot(x=x, y=y3, data=database)

plot_comparacao('mes', 'assinantes', 'aceleracao', 'aumento', assinantes, 'Análise de Assinantes da NewsLetter')
```
